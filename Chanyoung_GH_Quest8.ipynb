```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import auc
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
Xrt = X[:, 0:2]
model = LogisticRegression()
model.fit(Xrt, y)

X_train, X_test, y_train, y_test = train_test_split(Xrt, y)
y_score = y_score = model.fit(X_train, y_train).decision_function(X_test)

roc_auc = roc_auc_score(y_test, y_score)

fpr, tpr, thresholds = roc_curve(y_test, y_score)

plt.figure()
plt.plot(fpr, tpr, lw=2, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate = Recall')
plt.ylabel('True Positive Rate = Sensitivity')
plt.title('ROC curve')
plt.legend(loc="lower right")
plt.show()
```

```python
Xr = X[:, 0].reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(Xr, y)
y_score = y_score = model.fit(X_train, y_train).decision_function(X_test)

roc_auc = roc_auc_score(y_test, y_score)

fpr, tpr, thresholds = roc_curve(y_test, y_score)

Xt = X[:,1].reshape(-1,1)
Xt_train, Xt_test, yt_train, yt_test = train_test_split(Xt, y)
yt_score = yt_score = model.fit(Xt_train, yt_train).decision_function(Xt_test)

roc_auc_t = roc_auc_score(yt_test, yt_score)

fprt, tprt, thresholdst = roc_curve(yt_test, yt_score)


plt.figure()
plt.plot(fpr, tpr, lw=2, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot(fprt, tprt, lw=2, color='green', label='ROC curve (area = %0.2f)' % roc_auc_t)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate = Recall')
plt.ylabel('True Positive Rate = Sensitivity')
plt.title('ROC curve')
plt.legend(loc="lower right")
plt.show()
```

```python
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100], 'penalty': ['l1','l2']}

from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(model, param_grid, cv=5) 
grid_search.fit(Xrt, y)

print('최적의 파라미터: ', grid_search.best_params_)
print('최고 성능 모델: ', grid_search.best_estimator_)
print('최고 교차검증 점수: {:.2f}'.format(grid_search.best_score_))
```

```python
algo = grid_search.best_estimator_

X_train, X_test, y_train, y_test = train_test_split(Xrt, y)
y_score = y_score = algo.fit(X_train, y_train).decision_function(X_test)

roc_auc = roc_auc_score(y_test, y_score)

fpr, tpr, thresholds = roc_curve(y_test, y_score)

plt.figure()
plt.plot(fpr, tpr, lw=2, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate = Recall')
plt.ylabel('True Positive Rate = Sensitivity')
plt.title('ROC curve')
plt.legend(loc="lower right")
plt.show()
```
